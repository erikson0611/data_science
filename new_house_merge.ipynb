```python
import pandas as pd
df = pd.read_csv('./done_data/total_new_clean.csv')
print(df.columns)
print(df.shape)
df.head(1)
```

```python
area_df = pd.read_csv('./done_data/VarArea.csv')
area_df.head(1)
```

```python
df = df.merge(area_df, left_on='鄉鎮市區', right_on='Area')
print(df.shape)
df.head(10)
```

```python
# 整理台北市各區收入
# https://data.taipei/#/dataset/detail?id=1d6d31e2-7c1f-402e-9f5a-92d8a2b0d634
income_df = pd.read_csv('./data/fi00103y13ac.csv')
income_df = income_df[income_df['年別']=='108年']
income_df = income_df[['行政區','每戶人數[人]','每戶成年人數[人]','一、所得收入總計[元]','可支配所得[元]','消費支出[元]','儲蓄[元]','所得總額[元]']].set_index('行政區')
income_df = income_df.rename(columns={"每戶人數[人]": "每戶人數", "每戶成年人數[人]": "每戶成年人數", "一、所得收入總計[元]": "所得收入總計","可支配所得[元]":"可支配所得","消費支出[元]":"消費支出","儲蓄[元]":"儲蓄","所得總額[元]":"所得總額"})
income_df = income_df.iloc[1:,:]
income_df
```

```python
# 整理人口資料
# https://data.taipei/#/dataset/detail?id=08f12111-5856-4f15-bc13-ae14fed5bd9d
pop_df = pd.read_csv('./data/pp00003mc.csv', encoding = 'big5')
pop_df = pop_df[pop_df['年月底別']=='109年10月']
pop_df = pop_df.iloc[1:,:].set_index('行政區')
pop_df = pop_df[['土地面積[平方公里]','人口數/總計[人]','人口數/男[人]','人口數/女[人]','人口密度[人/平方公里]']]
pop_df = pop_df.rename(columns={"土地面積[平方公里]": "土地面積", "人口數/總計[人]": "總人口數", "人口數/男[人]": "男性人數","人口數/女[人]":"女性人數","人口密度[人/平方公里]":"人口密度"})
pop_df
```

```python
# 合併收入與人口結構表
import pandas as pd
merge_df = pd.concat([pop_df, income_df], axis = 1)
a = area_df.merge(merge_df, left_on='Area', right_on='行政區')
a = a.set_index('Area')
a.to_csv('merge_df.csv')  
a
```

```python
# 合併人口、收入到大表
df2 = df.merge(merge_df, left_on='鄉鎮市區', right_on='行政區')
df2[df['鄉鎮市區']=='大安區'].head(5) #確認
print(df2.columns)
```

### 平方公尺轉換成坪數

```python
df2['土地移轉總面積平方公尺'] = df2['土地移轉總面積平方公尺'] / 3.305785
df2['車位移轉總面積平方公尺'] = df2['車位移轉總面積平方公尺'] / 3.305785
df2['建物移轉總面積平方公尺'] = df2['建物移轉總面積平方公尺'] / 3.305785
df2['單價元平方公尺'] = df2['單價元平方公尺'] * 3.305785
df2 = df2.rename(columns={"土地移轉總面積平方公尺": "土地移轉總坪數", "車位移轉總面積平方公尺": "車位移轉總坪數", "建物移轉總面積平方公尺": "建物移轉總坪數","單價元平方公尺":"每坪單價"})
```

```python
# 無完工日期，交易年月日參考價值低
# 總價、車位價格與總價直接相關，需要移除
df2 = df2.drop(['交易年月日','總價元','車位總價元'], axis = 1)
df2.to_csv('./done_data/new_house_merge.csv', index = False)
df2.head(1)
```
