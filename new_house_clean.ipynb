```python
# 找出路徑下所有的資料夾名稱
import os
import pandas as pd
allFileList = os.listdir('./data')
file_list = []
for file in allFileList:
    if file.startswith('real_estate'):
        data_dir = './data/' + file + '/a_lvr_land_b.csv'
        file_list.append(data_dir)
len(file_list)
```

```python
# 確認預計的交易數量
# 要注意，部分資料夾沒有新成屋資料，要try catch
count = 0
for i in file_list:
    try:
        df = pd.read_csv(i)
        df = df[2:]
        count += len(df)
    except:
        print(i)
print(count)
```

```python
# 建立一個預備使用的空DataFrameresult_df = pd.read_csv(file_list[0])
result_df = df[0:0]
result_df
```

```python
# 合併所有資料到result_df
for i in file_list:
    try:
        df = pd.read_csv(i)
        df = df[2:]
        result_df = pd.concat([result_df, df])
    except:
        print(i)
print(result_df.shape)
result_df.head()
```

```python
# '交易標的'與後方交易筆棟數資料重複，所以drop
# '非都市土地使用分區','非都市土地使用編定','建築完成年月','備註'缺失值過多，drop
# '編號','土地區段位置建物區段門牌'無意義
result_df = result_df.drop(['土地區段位置建物區段門牌','交易標的','非都市土地使用分區','非都市土地使用編定','建築完成年月','備註','編號'],axis = 1)
print(result_df.shape)
print(result_df.columns)
result_df.head(1)
```

```python
# 將交易筆棟數分割成三個欄位
import re
build_list_split = []
build_list = result_df['交易筆棟數'].values.tolist()
rex = r'(\u571f\u5730){1}([0-9]*)(\u5efa\u7269){1}([0-9]*)(\u8eca\u4f4d){1}([0-9]*)'
for row in build_list:
    match = re.search(rex, row)
    build_list_split.append([match[2],match[4],match[6]])
build_df_split = pd.DataFrame(build_list_split, columns = ['土地','建物','車位'])
print(build_df_split.shape)
build_df_split
```

```python
# 將新的dateframe加入原有資料表，並且移除原有資料
result_df = result_df.reset_index(drop=True)
build_df_split = build_df_split.reset_index(drop=True)
result_df = pd.concat([result_df, build_df_split], axis = 1)
result_df = result_df.drop(['交易筆棟數'],axis = 1)
result_df.head(2)
```

```python
# 把車位類別做處理，避免dropna過多
print(result_df.shape)
result_df['車位類別'].fillna(value='0', inplace=True)
result_df = result_df.dropna()
print(result_df.shape)
```

```python
import re
floor_list_split = []
floor_list = result_df['移轉層次'].values.tolist()
floor_rex = r'(.*)(\u5c64)'
for row in floor_list:
    match = re.search(floor_rex, row)
    num = cn2an.cn2an(match[1])
    floor_list_split.append(num)
floor_df_split = pd.DataFrame(floor_list_split, columns = ['轉移樓層'])
print(floor_df_split.shape)
floor_df_split.head()
```

```python
# 將新的dateframe加入原有資料表，並且移除原有資料
result_df = result_df.reset_index(drop=True)
floor_df_split = floor_df_split.reset_index(drop=True)
result_df = pd.concat([result_df, floor_df_split], axis = 1)
result_df = result_df.drop(['移轉層次'],axis = 1)
print(result_df.shape)
result_df.head(2)
```

```python
# 處理建物型態，不要括號後面的東西
import re
type_list_split = []
type_list = result_df['建物型態'].values.tolist()
type_rex = r'(.*)(\(){01}'
for row in type_list:
    try:
        match = re.search(type_rex, row)
        type_list_split.append(match[1])
    except:
        type_list_split.append(row)
type_df_split = pd.DataFrame(type_list_split, columns = ['建物類型'])
print(type_df_split.shape)
type_df_split.head()
```

```python
# 將新的dateframe加入原有資料表，並且移除原有'建物型態'
result_df = result_df.reset_index(drop=True)
type_df_split = type_df_split.reset_index(drop=True)
result_df = pd.concat([result_df, type_df_split], axis = 1)
result_df = result_df.drop(['建物型態'],axis = 1)
print(result_df.shape)
result_df.head(2)
```

```python
# 將總表進行輸出，完全覆蓋模式
import csv
with open('./done_data/total_new_clean.csv', 'w', newline='') as csvfile2:
    writer = csv.writer(csvfile2)
    writer.writerows([result_df.columns]) # 加入feature name
    writer.writerows(result_df.values.tolist()) 
```
